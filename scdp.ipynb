## Análise dos Dados

```python
# Dados estatisticos do valor do bilhete
df['Valor Bilhete'].describe()
```

```python
# Quantidades de Reserva de Passagens Solicitadas por Órgãos
df['Nome do órgão solicitante da viagem'].value_counts()
```

```python
# Quantidade de Reserva de Passagens por órgão superior
df['Nome do órgão superior'].value_counts()
```

```python
# Qual a reserva que contém o bilhete mais caro?
df[df['Valor Bilhete']==df['Valor Bilhete'].max()]
```

```python
# Sacanagem, só pq eu sou de Santa Maria também
```

```python
# E qual a reserva com bilhete mais barato?
df[df['Valor Bilhete']==df['Valor Bilhete'].min()]
```

```python
# Importando biblioteca de gráficos para visulização dos dados
import seaborn as sns
```

```python
# Distribuição do valor da passagem
sns.distplot(df['Valor Bilhete'].dropna())
```

```python
# Quantidade de gasto de bilhete por órgão superior em R$ ordenado por maior valor
soma = df.groupby(['Nome do órgão superior'])[['Valor Bilhete']].sum() ## [[]] pra retornar DataFrame
soma.sort_values(by='Valor Bilhete', ascending=False)
```

```python
# São 180 órgãos superiores diferentes
soma.shape
```

```python
# Histograma dos dados anteriores
sns.distplot(soma['Valor Bilhete'].dropna())
```

```python
# Verifica-se que há uma maioria de valores acumulados na base, abaixo de R$ 500000, com alguns valores extremos
sns.boxplot(data=soma)
```

```python
# Verificar quantidade de empresas aéreas
df['Companhia Aérea'].value_counts()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Distribuição da quantidade de empresas aéreas
df['Companhia Aérea'].value_counts().plot.barh()
```

```python
# Valor médio gasto por empresa aérea
df.groupby('Companhia Aérea')['Valor Bilhete'].mean()
```

```python
# Soma dos valores de bilhete gasto por empresa aérea
df.groupby('Companhia Aérea')['Valor Bilhete'].sum()
```

```python
# Verificando órgãos que viajam só com uma empresa
empresas = df.groupby('Nome do órgão superior')[['Companhia Aérea']].nunique() ## [[]] para retornar um Dataframe
empresas[empresas['Companhia Aérea'] <= 1]  ## Uso o dataframe pra seleção aqui
```

```python
# órgãos que viajaram com todas as empresas
empresas[empresas['Companhia Aérea'] == empresas['Companhia Aérea'].max()]
```

```python
# distribuição dos valores por companhia aérea
sns.distplot(df[df['Companhia Aérea'].str.strip()=='GOL Linhas Aéreas Inteligentes']['Valor Bilhete'].dropna(), label='GOL')
```

```python
# distribuição dos valores por companhia aérea
sns.distplot(df[df['Companhia Aérea'].str.strip()=='AVIANCA S.A']['Valor Bilhete'].dropna(), label='AVIANCA S.A')
sns.distplot(df[df['Companhia Aérea'].str.strip()=='AZUL LINHAS AÉREAS BRASILEIRA']['Valor Bilhete'].dropna(), label='AZUL')
sns.distplot(df[df['Companhia Aérea'].str.strip()=='GOL Linhas Aéreas Inteligentes']['Valor Bilhete'].dropna(), label='GOL')
sns.distplot(df[df['Companhia Aérea'].str.strip()=='Gol OU Latam - Parceria com Terceiro']['Valor Bilhete'].dropna(), label='Parceiros')
sns.distplot(df[df['Companhia Aérea'].str.strip()=='LATAM AIRLINES BRASIL']['Valor Bilhete'].dropna(), label='LATAM')
sns.distplot(df[df['Companhia Aérea'].str.strip()=='MAP TRANSPORTES AEREOS LTDA - MAP']['Valor Bilhete'].dropna(), label='MAP')
plt.legend()
```

```python
# Distribuição dos valores dos Bilhetes por companhia aérea - Boxplot
sns.set(rc={'figure.figsize':(14, 6)})
plt.xticks(rotation=45)
sns.boxplot(x='Companhia Aérea', y='Valor Bilhete', data=df)
```
